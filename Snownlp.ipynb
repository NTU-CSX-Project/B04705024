```python
doc = []
docJieba = []
#summary 
sents = normal.get_sentences(text)
for sent in sents:
    words = seg.seg(sent)
    words = normal.filter_stop(words)
    wordswithJieba = jieba.lcut(sent)
    wordswithJieba = normal.filter_stop(wordswithJieba)
    doc.append(words)
    docJieba.append(wordswithJieba)
    
rank = textrank.TextRank(doc)
rank.solve()
rankJieba = textrank.TextRank(docJieba)
rankJieba.solve()
print("==沒有Jieba==")
for index in rank.top_index(5):
    print(sents[index])
print("==Jieba==")
for index in rankJieba.top_index(5):
    print(sents[index])

```
